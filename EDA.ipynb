```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flight_data = pd.read_csv('flights.csv')
```

```python
flight_data.head()
```

```python
flight_data.describe()
```

```python
flight_data.info()
```

```python
counts_month = flight_data['month'].value_counts()
print(counts_month)
```

```python
grouped_year = flight_data.groupby('year')['passengers'].sum()
print(grouped_year)
```

```python
grouped_month = flight_data.groupby('month')['passengers'].mean()
print(grouped_month)
```

```python
counts_year = flight_data['year'].value_counts()
print(counts_year)
```

```python
flight_data.isna().sum()
```

```python
flight_data.duplicated().sum()
```

```python
flights_data = flight_data
```

```python
flights_data.head()
```

```python
sns.set(style = 'whitegrid')
sns.boxplot(x='year', y='passengers', data = flights_data)
```

```python
monthly_passengers = flights_data.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(monthly_passengers)
```

```python
sns.lineplot(x='year', y='passengers', data=flights_data.groupby('year').sum())
```

```python
sns.lineplot(data=flights_data, x='month', y='passengers')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(x='month', y='passengers', data=flights_data)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(x = 'month', y = 'passengers', hue = 'year', data = flights_data.groupby(['year', 'month']).mean())
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot( data=flights_data, x='passengers', bins=10)
```

```python
sns.lineplot(x= flights_data.index, y='passengers', data=flights_data)
```
